## Tensor

In deep learning, a tensor is a multi-dimensional array used to store data. Tensors generalize the concept of scalars, vectors, and matrices to higher dimensions, making them fundamental in representing complex data structures in deep learning models.


Here are the different types of tensors based on their dimensions:


- Scalar (0-D tensor): A single number. For example, 3 or 5.7.

- Vector (1-D tensor): An array of numbers. For example, [1,2,3] or [0.5,0.1,0.4]. Vector would have a magnitude and direction.

- Matrix (2-D tensor): A 2-dimensional array of numbers. For example,

[1	2	3\
4	5	6]

- 3-D tensor: A 3-dimensional array of numbers. For example, a stack of matrices.

- Higher-dimensional tensors: Tensors with more than three dimensions. For example, a batch of images can be represented as a 4-D tensor with dimensions (batch size, height, width, channels).

In deep learning, tensors are used to represent inputs, outputs, and model parameters. They are manipulated and transformed using various operations (such as addition, multiplication, and convolution) in neural networks. Libraries like TensorFlow and PyTorch provide efficient ways to create, manipulate, and perform operations on tensors, making them essential tools for building and training deep learning models.

# Dot Product Matrix Multiplication

### `torch.mathmul()` function

matrix-multiply-b.svg

The "Dot Product" is where we multiply matching members, then sum up:

(1, 2, 3) • (7, 9, 11) = 1×7 + 2×9 + 3×11
    = 58\
(1, 2, 3) • (8, 10, 12) = 1×8 + 2×10 + 3×12
    = 64

```python
t1 = torch.tensor([[[1, 2, 3],[7, 9, 11]]])
t2 = torch.tensor([[[[7, 8],[9, 10], [11, 12]]]])

torch.matmul(t1, t2)
```

mm is an alias of matmul

### There are two main rules that matrix multiplication needs to satisfy:

1. **Inner dimensions** must match:
* `(3, 2) @ (3, 2)` won't work
* `(2, 3) @ (3, 2)` will work
* `(3, 2) @ (2, 3)` will work

2. Resulting matrix has the shape of **Outer Dimensions**
* `(2, 3) @ (3, 2) -> (2, 2)`
* `(3, 2) @ (2, 3) -> (3, 3)`

```python
torch.rand(3, 2).shape
```

```python
torch.matmul(torch.rand(3, 2), torch.rand(3, 2))
```

```python
torch.matmul(torch.rand(2, 3), torch.rand(3, 2))
```

```python
torch.matmul(torch.rand(3, 2), torch.rand(2, 3))
```

# Shape Errors

```python
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]])

tensor_B = torch.tensor([[7, 10],
                         [8, 11],
                         [9, 12]])

tensor_A.shape, tensor_B.shape
```

```python
torch.mm(tensor_A, tensor_B)
```

The two matrix does't follow Rule 1 ie., Inner dimensions must match.

To fix our tensor shape issues, we can manipulate the shape of one of our tensors using **transpose**

* A **transpose** switches the axes or dimensions of a given tensor

* tensor.T is syntax

```python
tensor_B
```

```python
tensor_B.shape
```

```python
tensor_B.T
```

```python
tensor_B.T.shape
```

```python
torch.mm(tensor_A, tensor_B.T)
```
